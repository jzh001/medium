```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
def getRoadBearings(location):
    G = ox.graph_from_place(location, network_type='drive')
    ox.bearing.add_edge_bearings(G)
    entropy = ox.bearing.orientation_entropy(G.to_undirected())
    # print(entropy)
    ox.bearing.plot_orientation(G.to_undirected(), title=location)
    return entropy
```

```python
places = [
    "Manhattan, New York, USA",
    "Pittsburgh, Pennsylvania",
    "Salt Lake City, Utah",
    "Chicago, Illinois",
    "Las Vegas, Nevada",
    "Austin, Texas",
    "Dayton, Ohio",
    "Atlanta, Georgia",
    "San Francisco, California",
    "Phoenix, Arizona",
    "Paris, France",
    "City of London",
    "Berlin, Germany",
    "Rome, Italy",
    "Milan, Italy",
    "Helsinki, Finland",
    "Vienna, Austria",
    "Manila, Philippines",
    "Xi'an, China",
    "New Delhi, India",
    "Ho Chi Minh City, Vietnam",
    "Kuala Lumpur, Malaysia",
    "Singapore",
]
```

```python
bearingEntropy = {place : getRoadBearings(place) for place in places}
bearingEntropy
```

```python
plt.barh(list(bearingEntropy.keys()), bearingEntropy.values())
```
