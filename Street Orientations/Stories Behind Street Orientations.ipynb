# Stories Behind Street Orientations

```python
import osmnx as ox
```

```python
def getRoadBearings(location):
    G = ox.graph_from_place(location, network_type='drive')
    ox.bearing.add_edge_bearings(G)
    ox.bearing.plot_orientation(G.to_undirected(), title=location)
```

```python
places = [
    "Salt Lake City",
    "Washington DC",
    "San Francisco, California",
    "Houston, Texas",
    "Chicago, Illinois",
    "Des Moines, Iowa",
    "Toronto, Canada",
    "Philadelphia, Pennsylvania",
    "Seattle, Washington",
    "Detroit, Michigan",
    "Pittsburgh, Pennsylvania",
    "New York City",
    "Boston, Massachusetts",
    "Manhattan, New York, USA",
    "Brooklyn, New York, USA",
    "Rome, Italy",
    "London, UK",
    "Paris, France",
    "Berlin, Germany",
    "City of London",
    "1er arrondissement, Paris",
    "Mitte, Berlin",
    "Xi'an, China",
    "Beijing, China",
    "Shenzhen, China",
]
```

```python
for place in places:
    getRoadBearings(place)
```
